# VERİ ÖN İŞLEME 

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("C:\\Users\\aysem\\Desktop\\Proje.xlsx")
print(data)
```

```python
data.info()
```

```python

data.groupby('Urun_Linki').size()
```

```python
#Aynı ürünlerin bilgileri olduğu için onları veri setinden temizledim
data.drop_duplicates(subset=['Urun_Linki'], inplace=True)
data
```

```python
#Sütun ismini değiştirdim
data.rename(columns={"Satici_Sayisi":"Baska_Satici_Sayisi"}, inplace=True)
```

```python
data.value_counts("Fiyat")
```

```python
data.drop("Fiyat", axis=1, inplace=True)
data
```

```python
data.drop("Kargo", axis=1, inplace=True)
```

```python
data
```

```python

data["Mevcut_Fiyat"]=data.Mevcut_Fiyat.str[:-3]
data.replace({"Favoriye Ekleyen Yok":0  },inplace=True)
data["Favori_Sayisi"]=data["Favori_Sayisi"].astype("int")
data.replace({"Değerlendirme Yok":0  },inplace=True)
data['Degerlendirme_Sayisi'] = data['Degerlendirme_Sayisi'].astype('int')
data.replace({"Satıcı Puanı Yok":0  },inplace=True)
data.replace({"Başka Satıcı Yok":0  },inplace=True)
data["Baska_Satici_Sayisi"]=data["Baska_Satici_Sayisi"].astype("int")
```

```python
data.info()
```

# VERİ ANALİZİ VE GÖRSELLEŞTİRME iŞLEMLERİ

```python
sns.scatterplot(x="Favori_Sayisi", y="Degerlendirme_Sayisi", data=data)
plt.title("Favori Sayısının Değerlendirme Sayısına Göre Dağılımı")
```

```python
d1=data[data["Favori_Sayisi" ]>0] 
d1
```

```python
sns.countplot(x="Baska_Satici_Sayisi", data=data)
plt.title("Başka Satıcı Sayısı")
```

```python
d2=data[data["Baska_Satici_Sayisi"].isin([1,2,3,4])]
d2
```

```python
d3=data[data["Kargo_Bilgisi"].isin(["Tahmini Kargoya Teslim: 8 gün içinde","Tahmini Kargoya Teslim: 6 gün içinde"])]
d3
```

```python
sonuc=pd.concat([d1,d2,d3])
sonuc
```

```python
sonuc.drop_duplicates(subset=['Urun_Linki'], inplace=True)
sonuc
```
